```
import librosa
import numpy as np
import math
import tensorflow as tf
import csv
from pypesq import pesq
import gc
import random

from IPython.display import Audio
```

```
# 開啟 CSV 檔案
def save_timit():
  with open('./timit/train_data.csv', newline='') as csvfile:
    path = ''
    timit = []
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
      if row[5][-3:] == "wav":
        path = row[5]
        sound = librosa.audio.load("./timit/data/{}".format(path),16000)[0]
        # sound -= np.mean(sound)
        # sound /= np.linalg.norm(sound)
        timit.append(sound)
    np.save('timit', timit)
```

```
save_timit()
```

```
def save_nonspeech():
  nonspeech = []
  for idx in range(100):
    sound = librosa.audio.load("./Nonspeech/n{}.wav".format(idx+1),16000)[0]
    # sound -= np.mean(sound)
    # sound /= np.linalg.norm(sound)
    nonspeech.append(sound)
  np.save('nonspeech', nonspeech)
```

```
save_nonspeech()
```

```
def to_noise_speech(speech, noise, snr):
  speech_len = speech.shape[0]
  noise_len = noise.shape[0]
  noise = np.repeat(noise, math.ceil(speech_len/noise_len), 0)
  noise = noise[:speech_len]
 
  from numpy.linalg import norm
  noise = noise / norm(noise) * norm(speech) / (10.0 ** (0.05 * snr))
  mix = speech + noise
 
  return mix
```
