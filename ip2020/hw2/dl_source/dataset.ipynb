```
from PIL import Image
import numpy as np
```

```
def crop_img(img,x,y):
  return img[64+y*64:320+y*64,64+x*64:320+x*64]
```

```
def rotate_img(img,rad):
  return np.array(Image.fromarray(img).rotate(rad))
```

```
def more_data(img):
  img = np.array(img)[64:448,64:448]
  imgs = []
  for x in [-1,0,1]:
    for y in [-1,0,1]:
      _img = crop_img(img,x,y)
      for rad in [0,90,180,270]:
        imgs.append(rotate_img(_img,rad))
  return imgs
```

```
data_num = [20, 21, 24, 21, 20, 22, 19, 21, 21, 18]

for i in range(len(data_num)):
  label = []
  MR = []
  for j in range(data_num[i]):
    label += zip(*[more_data(Image.open('{}/{}/{}.jpg'.format(i, t, j))) for t in ['CT','FT','MN']])
    MR += zip(*[more_data(Image.open('{}/{}/{}.jpg'.format(i, t, j))) for t in ['T1','T2']])
  np.savez("{}/data.npz".format(i), label=np.array(label), MR=np.array(MR),)
```
